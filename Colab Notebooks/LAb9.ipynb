```python
import numpy as np

def mean(data):
    n = len(data)
    mean = sum(data)/n
    return mean

def variance(data):
    n = len(data)
    mean = sum(data) / n
    deviation = [(x-mean) ** 2 for x in data]
    var = sum(deviation) / (n-1)
    return var

def stdev(data):
    import math
    var = variance(data)
    std_dev = math.sqrt(var)
    return std_dev

data = np.array([0,8,12,20])

print(stdev(data))
print(mean(data))
```

```python
import numpy as np
import scipy.linalg as la

A = np.array([[5, -2], [6, -2]])
print(A)
results = la.eig(A)
print("The eigenvalues are: ", results[0])
print("The eigenvectors are: ", results[1])
```

```python
import numpy as np
from numpy.linalg import eig

a = np.array([[0,2],[2,3]])
w,v = eig(a)
print("E-value:", w)
print("E-vector: ", v)
```

```python
import numpy as np

m = np.mat("3 -2;1 0")
print("Original Matrix\n", m)

w, v = np.linalg.eig(m)
print("E-value:", w)
print("E-vector: ", v)
```
