```python
n = 13
i = 2
prime = 1

while i!=n:
  if(n%i==0):
    prime = 0
  i = i+1

if(prime == 1):
  print("PRIME")
else:
  print("NOT PRIME")
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def printName(self, name):
        print(self.name)

class Student(Person):
    pass

p1 = Person("Argha", 20)
print(p1.name)
print(p1.age)
print(type(p1))

x = Student("Argha", 20)
x.printName("Argha")
```
