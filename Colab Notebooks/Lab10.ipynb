```python
import collections

def bubble_sort(collection):
    for i in range(len(collection) - 1):
        for j in range(len(collection) - i - 1):
            if(collection[j] > collection[j+1]):
                collection[j], collection[j+1] = collection[j+1], collection[j]

    return collection

my_list = [5, 3, 2, 1, 4]
sorted_list = bubble_sort(my_list)
print(sorted_list)
```

```python
# merge sort
import collections

def merge_sort(collection):
    if len(collection) <= 1:
        return collection

    mid = len(collection) // 2
    left = merge_sort(collection[:mid])
    right = merge_sort(collection[mid:])

    return merge(left, right)

def merge(left, right):
    res = collections.deque()
    i = 0
    j = 0
    while i < len(left) and j < len(right):
        if left[i]  <= right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1

    while i < len(left):
        res.append(left[i])
        i += 1

    while j < len(right):
        res.append(right[j])
        j += 1

    return list(res)

collection = [1,3,3,4,6,6,5,2,2,4,2,7,8,1,4,3]
sorted_collection = merge_sort(collection)
print(sorted_collection)
```

```python
# selection sort
import collections

def selection_sort(collection):
    for i in range(len(collection) - 1):
        min_index = i
        for j in range(i+1, len(collection)):
            if collection[j] < collection[min_index]:
                min_index = j

        collection[i], collection[min_index] = collection[min_index], collection[i]

    return collection

collection = [1,3,34,4,6,76725,452,512,2,4,12,87,85,1,434,3]
sorted_collection = merge_sort(collection)
print(sorted_collection)
```

```python
# insertion sort
import collections

def insertion_sort(collection):
    for i in range(1, len(collection)):
        curr = collection[i]
        j = i-1
        while j >= 0 and collection[j] > curr:
            collection[j+1] = collection[j]
            j -= 1
        collection[j+1] = curr

    return collection

collection = [1,3,34,4,6,76725,452,512,2,4,12,87,85,1,434,3]
sorted_collection = merge_sort(collection)
print(sorted_collection)
```

```python
# Binary Search Tree

class Node:
    def __init__(self, key):
        self.key = key
        self.left = None
        self.right = None

def inorder(root):
    if root is not None:
        inorder(root.left)
        print(str(root.key) + "->", end=' ')
        inorder(root.right)

def insert(node, key):
    if node is None:
        return Node(key)

    if key < node.key:
        node.left = insert(node.left, key)
    else:
        node.right = insert(node.right, key)

    return node

def minValueNode(node):
    current = node

    while(current.left is not None):
        current = current.left

    return current

def deleteNode(root, key):
    if root is None:
        return root

    if key < root.key:
        root.left = deleteNode(root.left, key)
    elif(key > root.key):
        root.right = deleteNode(root.right, key)
    else:
        if root.left is None:
            temp = root.right
            root = None
            return temp

        elif root.right is None:
            temp = root.left
            root = None
            return temp

        temp = minValueNode(root.right)
        root.key = temp.key
        root.right = deleteNode(root.right, temp.key)

    return root

root = None
root = insert(root, 8)
root = insert(root, 3)
root = insert(root, 1)
root = insert(root, 6)
root = insert(root, 7)
root = insert(root, 10)
root = insert(root, 14)
root = insert(root, 4)

print("Inorder traversal: ", end=' ')
inorder(root)

print("\nDelete 10")
root = deleteNode(root, 10)
print("Inorder traversal: ", end=' ')
inorder(root)
```
