```python
class Greeter(object):
    def __init__(self, name):
        self.name = name
    def greet(self, loud=False):
        if loud:
            print("HELLO, %s!" % self.name.upper())
        else:
            print("Hello, %s!" % self.name)

g = Greeter("Argha")
g.greet()
g.greet(loud=True)
```

```python
# map
print(list(filter(lambda x: (x%2 == 0), {1,5,4,6,8,11,3,12})))
print(filter(lambda x: (x%2 == 0), [1,5,4,6,8,11,3,12]))
```

```python
my_pets = ['alfred', 'tabitha', 'william', 'arla']
uppered_pets = list(map(str.upper, my_pets))
print(uppered_pets)

fruit = ['Apple', 'Mango', 'Pear']
map_obj = map(lambda s: s[0] == "A", fruit)
print(list(map_obj))
```

```python
# reduce
from functools import reduce
def add(x, y):
    return x+y
list = [2, 4, 7, 3]
print(reduce(add, list))

# using lambda function
print(reduce(lambda x, y: x+y, list))
print("With an initial value: " + str(reduce(lambda x,y: x+y, list, 11)))
```

```python
my_list = [[3,5,7,5],[1,2,9,4],[32,27,10,20]]
print(my_list)
sort_list = lambda num: (sorted(n) for n in num)
third_largest = lambda num, func: [l[len(l)-2] for l in func(num)]
result = third_largest(my_list, sort_list)
print(result)
```
