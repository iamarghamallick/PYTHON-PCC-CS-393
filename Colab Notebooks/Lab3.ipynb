```python
from functools import reduce
def add(x, y):
    return x+y
list = [1,2,3,4,5,6]
print(reduce(add, list))
```

```python
# code to print the third largest number from the list
my_list = [[1,2,3], [4,5,6], [7,8,9]]
print(my_list)

sorted_list = lambda num: (sorted(n) for n in num)
third_largest = lambda num, func: [l[len(l)-2] for l in func(num)]
result = third_largest(my_list, sorted_list)
print(result)
```

```python
import os
import threading
from threading import Thread
from time import sleep

# get id of the current process
print("ID of the running main program: {}".format(os.getpid()))

# print name of the main thread
print("Main thread name: {}".format(threading.main_thread().name))

# create two new thread
def task():
    print("Starting a task")
    sleep(2)
    print("Done")

t1 = Thread(target=task)
t2 = Thread(target=task)

# start the threads
t1.start()
t2.start()

# wait for threads to complete
t1.join()
t2.join()
```

```python
def add(x, y):
    return x+y

def calculate(func, x, y):
    return func(x,y)

result = calculate(add, 4, 6)
print(result)
```

```python
def make_pretty(func):
    def inner():
        print("I am decorated")
        func()
    return inner

@make_pretty
def ordinary():
    print("I am ordinary")

ordinary()
```

```python
import numpy as np
import matplotlib as mpl

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('X axis label')
plt.ylabel('Y axix label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show() # for graphic appereance
```

```python
import numpy as np
import matplotlib as mpl

x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
# make the first plot
plt.plot(x, y_sin)
plt.title('Sine')

plt.subplot(2,1,2)
# make the second plot
plt.plot(x, y_cos)
plt.title('Cosine')

plt.show() # for graphic appereance
```

```python
#ChatGPT
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

num = int(input("Enter a number: "))

if num < 0:
    print("Factorial is not defined fot negative numbers")
else:
    print("The factorial:", factorial(num))
```
