# **Linked List In Python**

```python
class Node:
    def __init__(self, data, next=None):
        self.data = data
        self.next = next

class LL:
    def __init__(self, head=None):
        self.head = head

    def add_node(self, data):
        new_node = Node(data)
        curr_node = self.head
        if curr_node is None:
            self.head = new_node
        else:
            while curr_node.next is not None:
                curr_node = curr_node.next
            curr_node.next = new_node

    def delete_node(self, data):
        if self.head is None:
            return
        new_node = Node(data)
        curr_node = self.head
        while curr_node.next.data is not new_node.data:
            curr_node = curr_node.next
        curr_node.next = curr_node.next.next

    def print_list(self):
        curr_node = self.head
        while curr_node is not None:
            print(curr_node.data,end=" -> ")
            curr_node = curr_node.next
        print("None")

if __name__ == "__main__":
    lst = LL()
    while True:
        print("** MAIN MENU **")
        print("Press 1 to Add Element")
        print("Press 2 to Delete Element")
        print("Press 3 to Display")
        print("Press 0 to EXIT")
        choise = int(input("Enter your choise"))

        if choise == 1:
            value = int(input("Enter element to be added : "))
            lst.add_node(value)
            print("CURRENT ELEMENTS IN THE NODE")
            lst.print_list()
        if choise == 2:
            value = int(input("Enter the element to be deleted : "))
            lst.delete_node(value)
            print("CURRENT ELEMENTS IN THE NODE")
            lst.print_list()
        if choise == 3:
            print("CURRENT ELEMENTS IN THE NODE")
            lst.print_list()
        if choise == 0:
            print("Good bye user!")
            break
        else:
            print("Wrong Choise!! PLEASE TRY AGAIN!!")
```

## **Doubly Linked List**

```python
class Node:
    def __init__(self, data, next=None, prev=None):
        self.data = data
        self.next = next
        self.prev = prev

class DLL:
    def __init__(self, head=None):
        self.head = head

    def add_node(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        curr_node = self.head
        while curr_node.next:
            curr_node = curr_node.next
        new_node.prev = curr_node

    def delete_node(self, data):
        if not self.head:
            return
        curr_node = self.head
        while curr_node:
            if curr_node.data == data:
                if curr_node.prev:
                    curr_node.prev.next = curr_node.next
                else:
                    self.head = curr_node.next
                if curr_node.next:
                    curr_node.next.prev = curr_node.prev
                return
            curr_node = curr_node.next

    def print_list(self):
        curr_node = self.head
        while curr_node:
            print(curr_node.data,end=" -> ")
            curr_node = curr_node.next
        print("None")

if __name__ == "__main__":
    lst = DLL()
    while True:
        print("** MAIN MENU **")
        print("Press 1 to Add Element")
        print("Press 2 to Delete Element")
        print("Press 3 to Display")
        print("Press 0 to EXIT")
        choise = int(input("Enter your choise"))

        if choise == 1:
            value = int(input("Enter element to be added : "))
            lst.add_node(value)
            print("CURRENT ELEMENTS IN THE NODE")
            lst.print_list()
        elif choise == 2:
            value = int(input("Enter the element to be deleted : "))
            lst.delete_node(value)
            print("CURRENT ELEMENTS IN THE NODE")
            lst.print_list()
        elif choise == 3:
            print("CURRENT ELEMENTS IN THE NODE")
            lst.print_list()
        elif choise == 0:
            print("Good bye user!")
            break
        else:
            print("Wrong Choise!! PLEASE TRY AGAIN!!")
```

```python
class BaseClass:
    def __init__(self, name):
        print("I am from __init__ BaseClass")
        self.name = name

    def say_hello(self):
        print("Hello, My name is", self.name)

class DerivedClass(BaseClass):
    def __init__(self, name, age):
        super().__init__(name)
        print("I am from __init__ DerivedClass")
        self.age = age

    def say_hello(self):
        print("Hello, My name is", self.name)

if __name__ == "__main__":
    b = BaseClass("Argha Mallick")
    d = DerivedClass("Argha Mallick", 20)
    d.say_hello()
```
