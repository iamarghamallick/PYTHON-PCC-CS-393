```python
def f(n):
    if n == 1 or n == 0:
        return 1
    return n * f(n-1)

num = 10
print(f"Factorial of {num} is {f(num)}")

# Iterative approch
def fact(n):
    f = 1
    for i in range(1, n+1):
        f = f * i
    return f

print(f"Factorial of {num} is {fact(num)} (using iteration)")
```

```python
def fibo(first, sec, n):
    if n == 0:
        return
    print(first, end=" ")
    fibo(sec, first+sec, n-1)

def fibonacci(n):
    first = 0
    sec = 1
    print(first, sec, end=" ")
    for i in range(n-2):
        next = first + sec
        first = sec
        sec = next
        print(next, end=" ")

num = 954
# fibo(0, 1, num)
print(f"{fibonacci(10)} (using iteration)")
```
