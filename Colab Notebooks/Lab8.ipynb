```python
import matplotlib.pyplot as plt
from scipy import stats

x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,70,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x,y)

def myFunc(x):
    return slope * x + intercept

mymodel = list(map(myFunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
```

```python
import time

def fibonacci(n):
    fib_seq = [0,1]
    if n <= 1:
        return fib_seq[:n+1]
    for i in range(2, n+1):
        next = fib_seq[-1] + fib_seq[-2]
        fib_seq.append(next)
    return fib_seq

n = int(input("Enter the number: "))
start = time.time()
if n<0:
    print("Please enter a positive number!")
else:
    res = fibonacci(n)
    print(f"Fibonacci sequence upto {n} is as follows")
    print(res)

end = time.time()
elapsed = end - start
print("Time taken", elapsed)
```
